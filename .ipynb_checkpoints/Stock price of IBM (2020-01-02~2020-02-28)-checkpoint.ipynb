```python
import pandas_datareader.data as web
```

```python
df = web.DataReader('IBM', 'stooq')
```

```python
#2020/01~2020/02的資料
df1 = df[:40]
print(df1)
```

```python
#開盤價
df1['Open'].plot()
```

```python
#收盤價
df1['Close'].plot()
```

```python
#成交量
df1['Volume'].plot()
```

```python
#開盤/收盤/當日最高/當日最低/成交量的敘述統計
df1.describe()
```
