### 挑戰題二

題目：防疫當道，大家都特別關心自身的健康狀況。然而，很多人的飲食習慣並不健康，因此有很多體重過輕或過重的情形。請做出一個BMI計算器並附上BMI對照表，讓使用者了解自己的BMI值是否在正常範圍內!

![BMI標準](https://scontent.ftpe4-1.fna.fbcdn.net/v/t1.0-9/83302659_2426649404331783_5249593098308157440_n.jpg?_nc_cat=107&_nc_sid=8024bb&_nc_ohc=6hiPjD2QBj8AX8Qt6LG&_nc_ht=scontent.ftpe4-1.fna&oh=1200e3833565c80e3d5ade8d6d4e1af4&oe=5E988156)

```python
def BMI(height,weight) : 
    s = weight/(height**2)
    return s
```

```python
one = input('身高(公尺)：')
two = input('體重(公斤)：')
'BMI為：'+ str(round(BMI(float(one),float(two)),4))
```
